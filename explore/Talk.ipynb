##Curve Fitting with Scikit-learn

- [scikit-learn](http://scikit-learn.org/) -- The library for machine learning in Python

```python
import sklearn.linear_model as lm
import sklearn.cross_validation as cv

x= artist_totals.familiarity.values
y= artist_totals.hotness.values

(x_train, x_test, y_train, y_test) = cv.train_test_split(x, y, test_size=.25)
lrp = lm.LinearRegression()
deg= 2
# Generate Vandermonde matrix
lrp.fit(np.vander(x_train, deg + 1), y_train)
y_fit= lrp.predict(np.vander(x_test, deg + 1))
plt.plot(x_train,y_train,'.',label='data')
plt.plot(x_test,y_fit,'or',label='fit')
plt.legend()
```

```python
lrp.score(np.vander(x_test, deg + 1),y_test)
```


## How about collaborative filtering?

```python

art1, art2= "Eminem", "Coldplay"

art1_listeners= ds[ds.artist_name=="Eminem"].uid.unique()
art2_listeners= ds[ds.artist_name=="Coldplay"].uid.unique()
common_listeners= set(art1_listeners).intersection(art2_listeners)
common_listeners
```

Based on the code by Martin O’Leary [http://mewo2.com/kaggle/2012/04/27/the-million-song-dataset-challenge-part-i/]

```python
import scipy.sparse
```

```python
max_song_id= ds.sid.max()
max_song_id
```

```python
colisten = scipy.sparse.lil_matrix((max_song_id, max_song_id))
by_song= ds[['uid', 'sid']].groupby('uid')

for name, group in by_song:
    for f in group['sid']:
        for t in group['sid']:
            colisten[f-1, t-1]+=1
colisten
```
