## Clean up data

## Plot

Introducing [matplotlib](http://matplotlib.org/), a Python plotting library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds['count'].hist(bins=100,log=True)
```

```python
artist_totals= ds[['artist_name', 'count', 'artist_hotttnesss', 'artist_familiarity']] \
.groupby('artist_name').agg([np.sum, np.min, np.max])
artist_totals= artist_totals[[0, 5, 8]]
artist_totals.columns= ['count', 'hotness', 'familiarity']
artist_totals= artist_totals.sort('count',ascending=False)
artist_totals
```

```python
artist_totals.plot(kind='scatter',x='familiarity',y='count')
```

```python
artist_totals.plot(kind='scatter',x='hotness',y='count')
```

```python
artist_totals.plot(kind='scatter',x='familiarity',y='hotness')
```

##Curve Fitting with Scikit-learn

- [scikit-learn](http://scikit-learn.org/) -- The library for machine learning in Python

```python
import sklearn.linear_model as lm
import sklearn.cross_validation as cv

x= artist_totals.familiarity.values
y= artist_totals.hotness.values

(x_train, x_test, y_train, y_test) = cv.train_test_split(x, y, test_size=.25)
lrp = lm.LinearRegression()
deg= 3
# Generate Vandermonde matrix
lrp.fit(np.vander(x_train, deg + 1), y_train)
y_fit= lrp.predict(np.vander(x_test, deg + 1))
plt.plot(x_train,y_train,'.',label='data')
plt.plot(x_test,y_fit,'or',label='fit')
plt.legend()
```

```python
# Let's see how good the fit was
lrp.score(np.vander(x_test, deg + 1),y_test)
```

```python
# Coefficients of the model, use this to implement model in production

lrp.coef_
```
